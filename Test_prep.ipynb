```python
import numpy as np 
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
df_students = pd.read_csv("students.csv")
df_teachers = pd.read_csv("teachers.csv")
df_courses = pd.read_csv("courses.csv")
df_student_course = pd.read_csv("studnet_course.csv")
```

```python
#מחק את כל הרשומות המכילות מידע חסר
df_students = df_students.dropna()
df_teachers = df_teachers.dropna()
df_courses = df_courses.dropna()
df_student_course = df_student_course.dropna()

#מחק את כל הרשומות הכפולות
df_students.drop_duplicates()
df_teachers.drop_duplicates()
df_courses.drop_duplicates()
df_student_course.drop_duplicates()
```

```python
#ספור כמה תלמידים יש מכל מגדר בטבלת students.csv
gender_counts = df_students['gender'].value_counts()
print(gender_counts)
```

```python
#הצג גם גרף עמודות (bar plot)
gender_counts.plot(kind='bar')

plt.title('Number of students by gender')
plt.xlabel('gender')
plt.ylabel('amount of students')

plt.tight_layout() 
plt.show()
```

```python
#צור היסטוגרמה להצגת התפלגות הגילאים של התלמידים

plt.hist(df_students['age'])
plt.grid(True)
plt.show()
```

```python
#ספר כמה קורסים יש בכל קטגוריה בטבלת courses.csv
courses_counts = df_courses['course_category'].value_counts()
print(courses_counts)
```

```python
#הצג גם גרף עמודות (bar plot)
courses_counts.plot(kind='bar')

plt.title('Number of courses by category')
plt.xlabel('category')
plt.ylabel('amount of courses')
plt.xticks(rotation=0) 
plt.tight_layout() 
plt.show()
```

```python
#מצא את העיר עם הכי הרבה תלמידים, ואת העיר עם הכי מעט תלמידים בטבלת students.csv
city_counts = df_students['city'].value_counts()

max_city = city_counts.idxmax()
min_city = city_counts.idxmin()
print(f'max:',max_city)
print(f'min:',min_city)
```

```python
#צור טבלה שמציגה, עבור כל קטגוריה, את שמות הקורסים הייחודיים השייכים אליה

courses_by_category = df_courses.groupby('course_category')['course_name'].unique().reset_index()

courses_by_category
```
